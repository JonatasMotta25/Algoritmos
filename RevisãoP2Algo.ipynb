```python
import os

def adicionar_aluno(alunos):
  try:
    nome = input("Digite o nome do aluno: ")
    while True:
      try:
        num_notas = int(input("Digite o número de notas (2 a 5): "))
        if 2 <= num_notas <= 5:
          break
        else:
          print("Número de notas inválido. Digite um valor entre 2 e 5.")
      except ValueError:
        print("Entrada inválida. Digite um número inteiro.")

    notas = []
    for i in range(num_notas):
      while True:
        try:
          nota = float(input(f"Digite a nota {i+1}: "))
          notas.append(nota)
          break
        except ValueError:
          print("Entrada inválida. Digite um número.")

    media = sum(notas) / len(notas)
    alunos.append({"nome": nome, "notas": notas, "media": media})
    print(f"Aluno {nome} adicionado com sucesso!")
  except Exception as e:
    print(f"Erro ao adicionar aluno: {e}")


def ordenar_alunos(alunos):
  try:
    alunos.sort(key=lambda aluno: aluno['media'], reverse=True)
    print("Alunos ordenados por média com sucesso!")
  except Exception as e:
    print(f"Erro ao ordenar alunos: {e}")


def salvar_em_arquivo(alunos, nome_arquivo="alunos.txt"):
  try:
    with open(nome_arquivo, "w") as arquivo:
      for aluno in alunos:
        arquivo.write(f"{aluno['nome']},{aluno['media']}\n")
    print(f"Dados dos alunos salvos em {nome_arquivo} com sucesso!")
  except Exception as e:
    print(f"Erro ao salvar dados em arquivo: {e}")

def exibir_alunos(alunos):
  for aluno in alunos:
    print(f"Nome: {aluno['nome']}")
    print(f"Média: {aluno['media']}")
    print("Notas:")
    for i, nota in enumerate(aluno['notas'], start=1):
      print(f"Nota {i}: {nota}")
    print()

alunos = []

while True:
  print("\nMenu:")
  print("1. Adicionar aluno")
  print("2. Ordenar alunos por média")
  print("3. Salvar dados em arquivo")
  print("4. Exibir alunos")
  print("5. Sair")

  opcao = input("Digite a opção desejada: ")

  if opcao == "1":
    adicionar_aluno(alunos)
  elif opcao == "2":
    ordenar_alunos(alunos)
  elif opcao == "3":
    salvar_em_arquivo(alunos)
  elif opcao == "4":
    exibir_alunos(alunos)
  elif opcao == "5":
    break
  else:
    print("Opção inválida.")


```

```python
import os

def adicionar_livro(livros):
    try:
        titulo = input("Digite o título do livro: ")
        autor = input("Digite o autor do livro: ")
        while True:
            try:
                ano = int(input("Digite o ano de publicação: "))
                if ano > 0:
                    break
                else:
                    print("Ano inválido. Digite um ano positivo.")
            except ValueError:
                print("Entrada inválida. Digite um número para o ano.")
        while True:
            try:
                paginas = int(input("Digite o número de páginas: "))
                if paginas > 0:
                    break
                else:
                    print("Número de páginas inválido. Digite um valor positivo.")
            except ValueError:
                print("Entrada inválida. Digite um número para as páginas.")
        livros.append({"titulo": titulo, "autor": autor, "ano": ano, "paginas": paginas})
        print(f"Livro '{titulo}' adicionado com sucesso!")
    except Exception as e:
        print(f"Erro ao adicionar livro: {e}")

def listar_livros(livros):
    if not livros:
        print("Nenhum livro cadastrado.")
        return
    for i, livro in enumerate(livros):
        print(f"\nLivro {i+1}:")
        print(f"Título: {livro['titulo']}")
        print(f"Autor: {livro['autor']}")
        print(f"Ano: {livro['ano']}")
        print(f"Páginas: {livro['paginas']}")

def ordenar_livros(livros):
    try:
        criterio = input("Ordenar por (ano/paginas): ").lower()
        ordem = input("Ordem (crescente/decrescente): ").lower()
        reverse = True if ordem == "decrescente" else False

        if criterio == "ano":
            livros.sort(key=lambda livro: livro['ano'], reverse=reverse)
        elif criterio == "paginas":
            livros.sort(key=lambda livro: livro['paginas'], reverse=reverse)
        else:
            print("Critério inválido.")
            return

        print("Livros ordenados com sucesso!")
    except Exception as e:
        print(f"Erro ao ordenar livros: {e}")


def salvar_livros(livros, nome_arquivo="biblioteca.txt"):
    try:
        with open(nome_arquivo, "w") as arquivo:
            for livro in livros:
                arquivo.write(f"{livro['titulo']},{livro['autor']},{livro['ano']},{livro['paginas']}\n")
        print(f"Dados dos livros salvos em {nome_arquivo} com sucesso!")
    except Exception as e:
        print(f"Erro ao salvar dados em arquivo: {e}")

def carregar_livros(nome_arquivo="biblioteca.txt"):
    try:
        livros = []
        if os.path.exists(nome_arquivo):
            with open(nome_arquivo, "r") as arquivo:
                for linha in arquivo:
                    dados = linha.strip().split(",")
                    if len(dados) == 4:
                        titulo, autor, ano, paginas = dados
                        livros.append({"titulo": titulo, "autor": autor, "ano": int(ano), "paginas": int(paginas)})
        return livros
    except Exception as e:
        print(f"Erro ao carregar livros do arquivo: {e}")
        return []

livros = []

while True:
    print ("Bem-vindo à Biblioteca Digital!")
    print("\nMenu:")
    print("1. Adicionar livro")
    print("2. Listar livros")
    print("3. Ordenar livros")
    print("4. Salvar livros em arquivo")
    print("5. Carregar dados do arquivo")
    print("6. Sair")

    opcao = input("Digite a opção desejada: ")


    if opcao == "1":
        adicionar_livro(livros)
    elif opcao == "2":
        listar_livros(livros)
    elif opcao == "3":
        ordenar_livros(livros)
    elif opcao == "4":
        salvar_livros(livros)
    elif opcao == "5":
        livros = carregar_livros()
    elif opcao == "6":
        break
    else:
        print("Opção inválida.")
print ("Bem-vindo à Biblioteca Digital!")
```

```python
import os

def adicionar_produto(produtos):
    try:
        nome = input("Digite o nome do produto: ")
        categoria = input("Digite a categoria do produto: ")
        while True:
            try:
                preco = float(input("Digite o preço do produto: "))
                if preco >= 0:
                    break
                else:
                    print("Preço inválido. Digite um valor não negativo.")
            except ValueError:
                print("Entrada inválida. Digite um número para o preço.")
        while True:
            try:
                quantidade = int(input("Digite a quantidade do produto: "))
                if quantidade >= 0:
                    break
                else:
                    print("Quantidade inválida. Digite um valor não negativo.")
            except ValueError:
                print("Entrada inválida. Digite um número para a quantidade.")
        produtos.append({"nome": nome, "categoria": categoria, "preco": preco, "quantidade": quantidade})
        print(f"Produto '{nome}' adicionado com sucesso!")
    except Exception as e:
        print(f"Erro ao adicionar produto: {e}")

def atualizar_quantidade(produtos):
    try:
        listar_produtos(produtos)
        while True:
            try:
              indice = int(input("Digite o número do produto a ser atualizado: ")) - 1
              if 0 <= indice < len(produtos):
                  break
              else:
                  print("Número de produto inválido.")
            except ValueError:
                print("Entrada inválida. Digite um número.")

        while True:
            try:
                alteracao = int(input("Digite a alteração na quantidade (positiva para aumentar, negativa para diminuir): "))
                nova_quantidade = produtos[indice]["quantidade"] + alteracao
                if nova_quantidade >= 0:
                    produtos[indice]["quantidade"] = nova_quantidade
                    print(f"Quantidade do produto '{produtos[indice]['nome']}' atualizada com sucesso!")
                    break
                else:
                    print("Quantidade resultante não pode ser negativa.")
            except ValueError:
                print("Entrada inválida. Digite um número.")
    except Exception as e:
        print(f"Erro ao atualizar quantidade: {e}")

def listar_produtos(produtos):
    if not produtos:
        print("Nenhum produto cadastrado.")
        return
    for i, produto in enumerate(produtos):
        print(f"\nProduto {i+1}:")
        print(f"Nome: {produto['nome']}")
        print(f"Categoria: {produto['categoria']}")
        print(f"Preço: {produto['preco']}")
        print(f"Quantidade: {produto['quantidade']}")

def ordenar_produtos(produtos):
    try:
        criterio = input("Ordenar por (preco/quantidade): ").lower()
        if criterio == "preco":
            produtos.sort(key=lambda produto: produto['preco'])
        elif criterio == "quantidade":
            produtos.sort(key=lambda produto: produto['quantidade'])
        else:
            print("Critério inválido.")
            return

        print("Produtos ordenados com sucesso!")
        listar_produtos(produtos)
    except Exception as e:
        print(f"Erro ao ordenar produtos: {e}")

def salvar_estoque(produtos, nome_arquivo="estoque.txt"):
    try:
        with open(nome_arquivo, "w") as arquivo:
            for produto in produtos:
                arquivo.write(f"{produto['nome']},{produto['categoria']},{produto['preco']},{produto['quantidade']}\n")
        print(f"Estoque salvo em {nome_arquivo} com sucesso!")
    except Exception as e:
        print(f"Erro ao salvar o estoque: {e}")

def carregar_estoque(nome_arquivo="estoque.txt"):
    produtos = []
    try:
        with open(nome_arquivo, "r") as arquivo:
            for linha in arquivo:
                nome, categoria, preco, quantidade = linha.strip().split(",")
                produtos.append({
                    "nome": nome,
                    "categoria": categoria,
                    "preco": float(preco),
                    "quantidade": int(quantidade)
                })
        print(f"Estoque carregado de {nome_arquivo} com sucesso!")
    except FileNotFoundError:
        print(f"Arquivo {nome_arquivo} não encontrado. Iniciando com estoque vazio.")
    except Exception as e:
        print(f"Erro ao carregar o estoque: {e}")
    return produtos

produtos = []

while True:
    print("Bem-vindo ao Sistema de Gestão de Estoque!")
    print("\nMenu:")
    print("1. Adicionar produto")
    print("2. Atualizar quantidade")
    print("3. Listar produtos")
    print("4. Ordenar produtos")
    print("5. Salvar estoque em arquivo")
    print("6. Carregar estoque do arquivo")
    print("7. Sair")

    opcao = input("Digite a opção desejada: ")
    if opcao == "1":
        adicionar_produto(produtos)
    elif opcao == "2":
        atualizar_quantidade(produtos)
    elif opcao == "3":
        listar_produtos(produtos)
    elif opcao == "4":
        ordenar_produtos(produtos)
    elif opcao == "5":
        salvar_estoque(produtos)
    elif opcao == "6":
        produtos = carregar_estoque()
    elif opcao == "7":
        break
    else:
        print("Opção inválida.")
```
